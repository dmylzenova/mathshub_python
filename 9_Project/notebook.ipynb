```python
# задаем URL-адрес API, ключ API и город для получения информации о погоде
url = 'https://api.openweathermap.org/data/2.5/weather'
api_key = '239f06dc1f50081036fc994468f58433'
city = 'Moscow'

# отправляем GET-запрос на URL-адрес API с заданными параметрами
params = {'q': city, 'appid': api_key, 'units': 'metric'}
response = requests.get(url, params=params)
```

Нам понадобятся регулярные выражения для того, чтобы найти нужно число в этой строке. Подробнее о регулярных выражениях по [ссылке](https://tproger.ru/translations/regular-expression-python/)

```python
import re
```

```python
# r'\d+,?\d*)\s*₽' ищет строку в данном формате, где
# \d+ - ЛЮБОЕ_КОЛИЧЕСТВО_ЦИФР_БОЛЬШЕ_1
# ,? опциональный знак вопроса 
# \d* - ЛЮБОЕ_КОЛИЧЕСТВО_ЦИФР_БОЛЬШЕ_0
# \s₽ - опциональный знак пробела
# ₽ - символ ₽

re.search(r'(\d+,?\d*)\s*₽', str(data[-1])).group(1).replace(',', '.')
```
