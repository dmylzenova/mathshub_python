````python
def delete_punctuation(s: str) -> str:
    c = ''
    for i in s:
        if i not in '?!.,:;-\"\'\\': # избегайте захардкоженных значений'
            c += i
    return c
````

# Ответы на вопросы:

- Инкапсуляция: Это принцип, по которому данные и методы, работающие с этими данными, объединяются в одном объекте. Таким образом, объекты скрывают свою внутреннюю реализацию от других объектов и предоставляют интерфейс для взаимодействия с ними. Пример: 

```
    class BankAccount:
    def __init__(self, balance):
        self.balance = balance
    
    def deposit(self, amount):
        self.balance += amount
    ...
```
    BankAccount инкапсулирует данные (баланс счета) и методы для работы с этими данными (внесение и снятие денег). Объект account предоставляет интерфейс для взаимодействия с балансом счета.

- Наследование: Это принцип, по которому класс может наследовать свойства и методы от другого класса. Таким образом, класс-потомок расширяет класс-родитель и наследует его функциональность. Пример:

```
class Human:
    def __init__(self, name):
        self.name = name
    
    def do_work(self):
        print("Do some work")

class Student(Human):
    def do_work(self):
        print("Do the 5th home work")
```
- Полиморфизм: Это принцип, по которому объекты разных классов могут иметь одинаковый интерфейс. Таким образом, методы, которые работают с объектами разных классов, могут вызывать один и тот же метод, но получать разные результаты. В примере выше метод do_work имеет одинаковый интерфейс, но будет давать разные результаты для объектов класса Human и Student.

- Абстракция - это принцип ООП, который позволяет скрыть детали реализации объекта и предоставить только интерфейс для работы с этим объектом.



````python
from abc import ABC, abstractmethod

class Shape(ABC):
    @abstractmethod
    def area(self):
        pass

class Rectangle(Shape):
    def __init__(self, width, height):
        self.width = width
        self.height = height
    
    def area(self):
        return self.width * self.height

class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius
    
    def area(self):
        return 3.14 * self.radius ** 2
````

- Что такое self
        
        В Python ключевое слово self используется для ссылки на текущий экземпляр класса. Когда мы создаем объект класса, Python автоматически передает этот объект в качестве первого аргумента каждому методу класса, который мы вызываем. По умолчанию этот аргумент называется self.

        Чтобы получить доступ к атрибутам или методам класса внутри метода класса, мы должны использовать self перед именем атрибута или метода.

````python
class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius
    
    def area(self):
        return 3.14 * self.radius ** 2
````

- Что такое @staticmethod
        
        Использем, когда методу не нужен доступ к другим атрибутам и методам.

````python
class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def add_numbers(self, x, y):
        return x + y
    
    def area(self):
        return 3.14 * self.radius ** 2
    
    def some_method(self, x):
        return self.add_numbers(self.radius, x)
````

- @classmethod
    Использем, когда методу нужен доступ только к атрибутам класса (не объекта).

````python
class BaseModel:
    @classmethod                 # Класс-метод класса -> вместо экземпляра self передается класс cls
    def foo(cls):
        print(f"I was called from {cls.__name__}")

class RegressionModel(BaseModel):
    pass

class Classification(BaseModel):
    pass
````

````python
a = RegressionModel()
b = Classification()
a.foo()
````

````python
b.foo()
````

````python
class BaseModel:
    def foo(self):
        print(f"I was called from {self.__class__.__name__}")

class RegressionModel(BaseModel):
    pass

class Classification(BaseModel):
    pass

a = RegressionModel()
b = Classification()
a.foo()
````

- Наследование

````python
import math

class BaseModel:
    def __init__(self, weights, bias):
        self.weights = weights
        self.bias = bias
    
    def compute_linear_function(self, x):
        return self.weights * x + self.bias
    
class RegressionModel(BaseModel):
    def compute_result(self, x):
        result = self.compute_linear_function(x)
        return result
        
class ClassificationModel(BaseModel):
    @staticmethod
    def compute_sigmoid(x):
        return 1. / (1 + math.exp(-x))

    def compute_result(self, x):
        result = self.compute_linear_function(x)
        return int(self.compute_sigmoid(result) > 0.5)
````

````python
reg_model = RegressionModel(5.1, 0.2)
clf_model = ClassificationModel(5.1, 0.2)
reg_model.compute_result(1), clf_model.compute_result(1)
````

Кэширование

````python
def expensive_function(x):
    # Выполнение дорогостоящих вычислений
    result = x ** 2 + 2 * x + 1
    return result

cache = {}

def cached_function(x):
    if x in cache:
        print('use cached value')
        return cache[x]
    result = expensive_function(x)
    cache[x] = result
    return result

print(cached_function(2))  
````

````python
print(cached_function(2))  # (результат берется из кэша)
````
