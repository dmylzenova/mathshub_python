### 1. Подсчет расстояния между векторами (одномерными массивами).

```python
x = np.random.uniform(0, 1, (128))
y = np.random.uniform(0, 1, (128))
```

```python
def compute_dictance(x, y):
    res = []
    for (x_i, y_i) in zip(x, y):
        res.append((x_i - y_i) ** 2)
    return math.sqrt(sum(res))
```

```python
np.sqrt(((x - y) ** 2).sum())
```

```python
np.sqrt(((x - y) ** 2).sum()) - compute_dictance(x, y)
```

```python
np.isclose(compute_dictance(x, y), np.sqrt(((x - y) ** 2).sum()))
```

```python
np.linalg.norm((x - y))
```

```python
np.isclose(np.linalg.norm((x - y)), np.sqrt(((x - y) ** 2).sum()))
```

### 5. Stacking нескольких массивов в один массив (разные размерности)
в этом случае есть два способа 
- cropping (вырезание) - отрезать от каждого массива, чтобы сделать одинаковую размерность.
- padding (набивка) добавление дополнительных значений по краям.


```python
shapes = np.arange(5, 11)
np.random.choice(shapes)
```

```python
list_of_arrays = [np.random.random((np.random.choice(shapes), np.random.choice(shapes))) for x in range(5)]
```

```python
[x.shape for x in list_of_arrays]
```

```python
np.stack(list_of_arrays)
```

```python
min([x.shape[0] for x in list_of_arrays]), min([x.shape[1] for x in list_of_arrays])
```

```python
max_shapes = max([x.shape[0] for x in list_of_arrays]), max([x.shape[1] for x in list_of_arrays])
```

```python
?np.pad
```

```python
padded = [np.pad(x, ((0, max_shapes[0] - x.shape[0]), (0, max_shapes[1] - x.shape[1]))) for x in list_of_arrays]
```

```python
result = np.stack(padded)
result.shape
```

# Pandas

- [**Чтение и запись данных:**](###-**Чтение-и-запись-данных**) Pandas может использоваться для чтения и записи данных в различных форматах, включая CSV, Excel, JSON и SQL. Примеры задач: чтение данных из CSV-файла, запись данных в Excel-файл.

- **Фильтрация данных:** Pandas предоставляет инструменты для фильтрации данных на основе условий. Примеры задач: выборка строк, которые соответствуют определенному условию, выборка столбцов по условию.

- **Преобразование данных:** Pandas может использоваться для преобразования данных, таких как изменение типа данных, заполнение пропущенных значений и создание новых столбцов на основе существующих данных. Примеры задач: изменение типа данных столбца, заполнение пропущенных значений в столбце, создание нового столбца на основе существующих столбцов.

- **Группировка данных:** Pandas может использоваться для группировки данных и агрегирования информации по группам. Примеры задач: группировка данных по столбцу и подсчет среднего значения или медианы в каждой группе.


- **Объединение данных:** Pandas может использоваться для объединения данных из нескольких источников в один объект DataFrame. Примеры задач: объединение двух таблиц по общему столбцу, добавление столбца из одной таблицы в другую таблицу.

- **Анализ статистических данных:** Вычисление различных статистических параметров, таких как среднее значение, медиана, минимум, максимум и т.д.

- [**Визуализация данных:**](###--**Визуализация-данных**) Pandas может использоваться для визуализации данных с помощью библиотеки matplotlib и seaborn. Примеры задач: построение гистограммы распределения данных, построение диаграммы рассеяния.

- **Анализ временных рядов:** Pandas предоставляет инструменты для работы с временными рядами, включая ресемплирование, сглаживание и анализ трендов. Примеры задач: анализ временных рядов на основе ежедневных данных, ресемплирование данных для анализа еженедельных или ежемесячных трендов.

- **Машинное обучение:** Обучение моделей машинного обучения на основе данных в Pandas DataFrame, таких как регрессия, классификация, кластеризация и т.д. (с помощью сторонних библиотек)

### Производство солнечной энергии за период

```python
start, end = '2017-01', '2017-06'

fig, ax = plt.subplots()
ax.plot(energy_data.loc[start:end, 'Solar'],
        marker='.', linestyle='-', linewidth=0.5, label='Daily')
ax.plot(energy_data_mean.loc[start:end, 'Solar'],
        marker='o', markersize=8, linestyle='-', label='Weekly Mean Resample')
ax.set_ylabel('Solar Production (GWh)')
ax.legend()
```

```python
energy_data_7d = energy_data[data_columns].rolling(7, center=True).mean()
energy_data_7d.sample(4)
```

```python
start, end = '2017-01', '2017-06'

fig, ax = plt.subplots()
ax.plot(energy_data.loc[start:end, 'Solar'],
        marker='.', linestyle='-', linewidth=0.5, label='Daily')
ax.plot(energy_data_mean.loc[start:end, 'Solar'],
        marker='o', markersize=8, linestyle='-', label='Weekly Mean Resample')
ax.set_ylabel('Solar Production (GWh)')

ax.plot(energy_data_7d.loc[start:end, 'Solar'],
        marker='.', linestyle='-', label='7-d Rolling Mean')
ax.legend()
```

### Тренд потребления энергии
Чтобы увидеть тренд, нужно аггрегировать данные за больший период, например, за год.

```python
energy_data_365d = energy_data[data_columns].rolling(window=365, center=True, min_periods=360).mean()
```

```python
import matplotlib.dates as mdates
```

```python
fig, ax = plt.subplots(figsize=(9, 5))
ax.plot(energy_data['Consumption'], marker='.', markersize=2, color='0.6',
        linestyle='None', label='Daily')
ax.plot(energy_data_7d['Consumption'], linewidth=2, label='7-d Rolling Mean')
ax.plot(energy_data_365d['Consumption'], color='0.2', linewidth=3,
        label='Trend (365-d Rolling Mean)')

# Set x-ticks to yearly interval and add legend and labels
ax.xaxis.set_major_locator(mdates.YearLocator())
ax.legend()
ax.set_xlabel('Year')
ax.set_ylabel('Consumption (GWh)')
ax.set_title('Trends in Electricity Consumption');
```
